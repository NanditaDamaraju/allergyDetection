```python
%matplotlib inline
import cv2
from skimage.filters import prewitt
from skimage.morphology import remove_small_objects
from scipy.signal import find_peaks
import matplotlib.pyplot as plt
import numpy as np
```

```python
def chop_image(image_path):
    """Chop images and plot them on a grid
    
    Inputs
    - image_path: path to the image file in .jpg format
    
    """
    image_rgb = cv2.imread(image_path)
    image = cv2.imread(image_path,0)
    im1 = cv2.bilateralFilter(image, 40, 500, 500)
    im2  = prewitt(im1)
    im3 = im2 >0.0025
    im4 = remove_small_objects(im3,min_size=100)
    im5 = np.int8(im4)
    
    horizontal = np.sum(im5,axis=0)
    vertical = np.sum(im5, axis = 1)
    
    h_pos, h_heights = find_peaks(horizontal, height = 1000, distance = 500)
    v_pos, v_heights = find_peaks(vertical, height = 1000, distance = 500)
    fig, axes = plt.subplots( len(v_pos)-1,len(h_pos)-1)

    for h in range(len(h_pos)-1) :
        for v in range(len(v_pos)-1):
            
            start_h = h_pos[h]
            end_h = h_pos[h+1]

            start_v = v_pos[v]
            end_v = v_pos[v+1]

            axes[v,h].imshow(image_rgb[start_v:end_v,start_h:end_h, :])
```

```python
chop_image('009-1982-5.jpg')
```
